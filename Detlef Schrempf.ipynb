#### Following my NBA 3 Pointer analysis (https://crawstat.com/2020/05/27/the-nba-3-pointers-staggering-rise/), a reader asked how Detlef Schrempf stacks up. As Schrempf is one of my all-time favorite shooters, I was curious to find out myself. 

#### Schrempf was a small forward for the Seattle SuperSonics and a lethal 3 point (and mid-range) shooter complementing Shawn Kemp and Gary Payton during the team's storied '90s run. He was a 3-time All Star and and 2-time 6th Man of the Year. He also played for the Dallas Mavericks, Indiana Pacers, and Portland Trailblazers. Sidenote: he made a cameo in three episodes of one of my favorite shows Parks and Recreation. 

#### Let's take a look at how Schrempf stacks up from 3 point range. Notice that for his career, while he was fairly in line with the league in 3PAr, he was leaps and bounds above in 3PA, 3P%, and 3P. What made him a such an efficient shooter was his 3P% (and FG% and FT%). For example, in his 1994-1995 Sonics campaign, his FG%/3P%/FT% was a deadly 52/51/84.  

```python
# Break out Detlef Schrempf

schrempf = nba1.loc[nba1.Player == "Detlef Schrempf", :]

# Lineplot of median 3PA per season by year for league and Schrempf 

plt.figure(dpi=100)
sns.lineplot(x="Year", y="3PA", estimator = np.median, ci = None, data=nba1)
sns.lineplot(x="Year", y="3PA", ci = None, data=schrempf)
plt.ylabel("3 Point Attempts Per Season")
plt.title("Median 3 Point Attempts from 1980 to 2017")
plt.legend(labels=["NBA","Schrempf"])
plt.show()
```

```python
# Lineplot of median 3PAr per season by year for league and Schrempf. 

plt.figure(dpi=100)
sns.lineplot(x="Year", y="3PAr", estimator = np.median, ci = None, data=nba1)
sns.lineplot(x="Year", y="3PAr", ci = None, data=schrempf)
plt.ylabel("3 Point Attempt Rate Per Season")
plt.title("Median 3 Point Attempt Rate from 1980 to 2017")
plt.legend(labels=["NBA","Schrempf"])
plt.show()
```

```python
# Lineplot of median 3P% per season by year for league and Schrempf. 

plt.figure(dpi=100)
sns.lineplot(x="Year", y="3P%", estimator = np.median, ci = None, data=nba1)
sns.lineplot(x="Year", y="3P%", ci = None, data=schrempf)
plt.ylabel("3 Point %  Per Season")
plt.title("Median 3 Point % from 1980 to 2017")
plt.legend(labels=["NBA","Schrempf"])
plt.show()
```

```python
# Lineplot of median 3P per season by year for league and Schrempf. 

plt.figure(dpi=100)
sns.lineplot(x="Year", y="3P", estimator = np.median, ci = None, data=nba1)
sns.lineplot(x="Year", y="3P", ci = None, data=schrempf)
plt.ylabel("3 Pointers Made  Per Season")
plt.title("Median 3 Pointers Made from 1980 to 2017")
plt.legend(labels=["NBA","Schrempf"])
plt.show()
```
